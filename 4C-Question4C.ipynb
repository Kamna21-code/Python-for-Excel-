```python
import pandas as pd
```

```python
data1 = pd.read_excel(r'PatientData_Programming Assignment.xlsx',sheet_name = 'Sheet1')
data2 = pd.read_excel(r'PatientData_Programming Assignment.xlsx',sheet_name = 'Sheet2')
```

```python
data =  data1.merge(data2, how='inner', on='Patient_ID', left_index=True, right_index=True, copy=True)
data.head()
```

```python
data['HR'] = 0
for i in range(data.shape[0]):
    if (data.PR_positive.loc[i] == 1) or (data.ER_positive.loc[i] == 1):
        data['HR'][i] = 1
        
data.head()
```

```python
data['HR'] = 0
data['HER2(+)_HR(+)'] = 0
data['HER2(+)_HR(-)'] = 0
data['HER2(-)_HR(+)'] = 0
data['HER2(-)_HR(-)'] = 0
for i in range(data.shape[0]):
    if (data.HER2_positive.loc[i] == 1) and (data.HR.loc[i] == 1):
        data['HER2(+)_HR(+)'][i] = 1
    elif (data.HER2_positive.loc[i] == 1) and (data.HR.loc[i] == 0):
        data['HER2(+)_HR(-)'][i] = 1
    elif (data.HER2_positive.loc[i] == 0) and (data.HR.loc[i] == 1):
        data['HER2(-)_HR(+)'][i] = 1
    elif (data.HER2_positive.loc[i] == 0) and (data.HR.loc[i] == 0):
        data['HER2(-)_HR(-)'] = 1
```

```python
rA = data.groupby('HER2(+)_HR(+)').sum('Length')
rB = data.groupby('HER2(+)_HR(-)').sum('Length')
rC = data.groupby('HER2(-)_HR(+)').sum('Length')
rD = data.groupby('HER2(-)_HR(-)').sum('Length')
```

```python
rA.to_excel('Question4CA.xlsx')
rB.to_excel('Question4CB.xlsx')
rC.to_excel('Question4CC.xlsx')
rD.to_excel('Question4CD.xlsx')
```
