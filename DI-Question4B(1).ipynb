```python
import pandas as pd
```

```python
data1 = pd.read_excel(r'PatientData_Programming Assignment.xlsx',sheet_name = 'Sheet1')
data2 = pd.read_excel(r'PatientData_Programming Assignment.xlsx',sheet_name = 'Sheet2')
```

```python
data =  data1.merge(data2, how='inner', on='Patient_ID', left_index=True, right_index=True, copy=True)
data.head()
```

```python
df1 = data[data.drug_390_admin_flag == 1].copy()
df2 = data[data.drug_390_admin_flag != 1].copy()
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.DataFrame()
df['Patient_id_who_take_390'] = df1.Patient_ID
df['Length_of_treatment_for_390_patient'] = df1.Length
df['Length_of_treatment_for_generic_Patient'] = df2.Length
df['Patient_id_who_are_generic'] = df2.Patient_ID
df.head()
```

```python
df1 = df1[['Patient_ID', 'Length']]
df2 = df2[['Patient_ID', 'Length']]
df1.to_excel('Patient_ID_and_Length_of_Patient_who_take_390.xlsx')
df2.to_excel('Patient_ID_and_Length_of_Patient_who_are_generic.xlsx')
```

```python
r1 = df1.groupby('Patient_ID').sum('Length')
r2 = df2.groupby('Patient_ID').sum('Length')
```
